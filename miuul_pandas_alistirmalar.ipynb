GÖREV 23

```python
mean_male = df[df["sex"]=="Male"].agg({"total_bill" : "mean"})
```

```python
mean_female = df[df["sex"]=="Female"].agg({"total_bill" : "mean"})
```

```python
def bill_flagger(sex, bill):
    if sex == "Male" and bill < mean_male[0]:
        return 0
    elif sex == "Male" and bill > mean_male[0]:
        return 1
    elif sex == "Female" and bill < mean_female[0]:
        return 0
    elif sex == "Female" and bill > mean_female[0]:
        return 1
```

```python
df["total_bill_flag"] = df.apply(lambda x: bill_flagger(x.sex, x.total_bill), axis=1)
```
